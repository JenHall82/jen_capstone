```python
counties.ste_name = counties.ste_name.explode()
counties
```

```python
tn_counties = counties.loc[(counties['ste_name'] == 'Tennessee')]
tn_counties.to_csv('../data/tn_counties.csv')
```

```python
tn_counties.coty_name_long = tn_counties.coty_name_long.explode()
tn_counties.to_json
```

```python
dist_scores_percent = pd.read_csv('../data/dist_scores_percent.csv')
dist_scores_percent
```

```python
dist_scores_percent = pd.merge(dist_scores_percent, tn_counties, left_on = 'County Name', right_on = 'coty_name_long', how = 'inner')
dist_scores_percent.reset_index()
```

```python
dist_scores_percent = dist_scores_percent.drop(columns = ['Unnamed: 0.1', 'ste_code', 'coty_code', 'coty_name', 'coty_area_code', 'coty_fp_code', 
                                                          'coty_gnis_code'])
dist_scores_percent
```

```python
scores_geo = gpd.GeoDataFrame(dist_scores_percent, 
                           crs = counties.crs, 
                           geometry = dist_scores_percent['geometry'])
```

```python
scores_geo.to_csv('../data/scores_geo.csv')
```
